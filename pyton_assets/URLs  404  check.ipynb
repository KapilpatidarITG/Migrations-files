```python
import csv
import requests
import pandas as pd
```

```python
def check_image_urls(csv_file):
    with open(csv_file, 'r') as file:
        reader = csv.DictReader(file)
      
        for row in reader:
            image_url = row['ImageURL']
            status = check_url_status(image_url)
            print(f"{image_url}: {status}")
            
def check_url_status(url):
    try:
        response = requests.head(url, timeout=10)
        return "CHalu hai" if response.status_code == 200 else "Inactive"
    except requests.ConnectionError:
        return "Inactive"

if __name__ == "__main__":
    csv_file_path = 'Book2.csv'
    check_image_urls(csv_file_path)
```
